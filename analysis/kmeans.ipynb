```python
n_clusters = 40
date_block_cutoff = 30
eval_lags = np.arange(1,25)
eval_lags = [ 1, 2, 3, 6, 12 ]

kmr = KMeansRegression( kg, n_clusters=n_clusters, preprocess=True, \
                       eval_lags=eval_lags, date_block_cutoff=date_block_cutoff, \
                       forecast_method='prophet' )
```

```python
kmr.fit_train()
kmr.predict_train(clip_forecasts=True)

print( 'Validation RMSE: {}'.format( rmse( kmr.yhat[TRAIN], kmr.yy[TRAIN] ) ) )
```

```python
kmr.predict_validation(clip_forecasts=True)

print( 'Validation RMSE: {}'.format( rmse( kmr.yhat[VALID], kmr.yy[VALID] ) ) )
```
